# **Extraindo várias receitas**

---



```python
import time
import random
```

```python
#numero_paginas = 150
numero_paginas = 30

links_receitas = []

for p in range(numero_paginas):
    url = f"https://www.receitadahora.com/bolos/page/{p}"
    response = requests.get(url, headers=headers)

    print(p, response)

    soup = BeautifulSoup(response.text, "html.parser")
    bloco = soup.find("div", id="meio-conteudo")

    links = [a['href'] for a in bloco.find_all('a', href=True)]
    links = [link for link in links if "/bolos/bolo-" in link]

    links_receitas += links

    # Esperar alguns segundos antes de ir para a próxima página
    tempo_espera = random.uniform(0, 0.1)
    time.sleep(tempo_espera)
```

# **Salvando os arquivos**

---



```python
import json
import pathlib
import re

data_dir = pathlib.Path("data")  # Diretório onde os arquivos serão salvos
data_dir.mkdir(exist_ok=True)

for i, link in enumerate(links_receitas):

    full_url = link  # Construindo a URL completa
    response = requests.get(full_url, headers=headers)

    soup = BeautifulSoup(response.text, "html.parser")

    title = soup.find("meta", property="og:title")["content"]
    description = soup.find("meta", attrs={"name": "description"})["content"]
    date = soup.find("meta", property="article:published_time")["content"]

    texto_materia = soup.find("div", id="meio-conteudo")
    texto = texto_materia.get_text(separator="\n").strip()

    dados = {
        "url": full_url,
        "titulo": title,
        "descricao": description,
        "texto": texto,
        "data": date,
    }



    # Removendo \n do texto
    texto_limpo = dados['texto'].replace('\n', ' ').strip()

    # Extraindo o tempo estimado de preparo
    tempo_preparo_match = re.search(r'Tempo de preparo:\s*(.*?)\s*Serve:', texto_limpo)
    dificuldade_match = re.search(r'Dificuldade:\s*(.*?)\s* ', texto_limpo)
    ingredientes_match = re.search(r'INGREDIENTES\s*(.*?)\s* MODO', texto_limpo)

    tempo_preparo = tempo_preparo_match.group(1) if tempo_preparo_match else None
    dificuldade = dificuldade_match.group(1) if dificuldade_match else None
    ingredientes = ingredientes_match.group(1) if ingredientes_match else None

    # Atualizando os dados
    dados['tempo_preparo'] = tempo_preparo
    dados['dificuldade'] = dificuldade
    dados['ingredientes'] = ingredientes


    filename = f"receita_{i}.json"

    file_path = data_dir / filename

    with file_path.open("w", encoding="utf-8") as f:
        json.dump(dados, f, ensure_ascii=False, indent=4)

    print(f"Arquivo salvo: {file_path}")
    tempo_espera = random.uniform(0, .5)
    time.sleep(tempo_espera)
```

# **Análise dos arquivos**

A seguir vamos ver as frequências de tempo, dificuldade e ingredientes, a fim de entender como se comportam as receitas

## 1 - dificuldade

```python
import json
import pathlib
import matplotlib.pyplot as plt

# Caminho da pasta com os arquivos JSON
data_dir = pathlib.Path("data")

# Contador
dificuldade_counter = {}

grouped_dif = {
    "média": ["médio", "média"]  # Agrupar médio e média em um só
}

# Abrindo e processando os arquivos JSON
for json_file in data_dir.rglob("*.json"):
    with json_file.open("r", encoding="utf-8") as f:
        data = json.load(f)
        dificuldade = data.get("dificuldade")
        if dificuldade:
            dificuldade = dificuldade.lower()  # Converte para minúsculas

            # Agrupando "médio" e "média"
            for grouped in grouped_dif:
                if dificuldade in grouped_dif[grouped]:
                    dificuldade = grouped


            if dificuldade in dificuldade_counter:
                dificuldade_counter[dificuldade] += 1
            else:
                dificuldade_counter[dificuldade] = 1


print("Contagem de dificuldades:")
for dificuldade, count in dificuldade_counter.items():
    print(f"{dificuldade}: {count}")

# Criando um gráfico de barras
plt.bar(dificuldade_counter.keys(), dificuldade_counter.values(), color="skyblue")
plt.xlabel("Dificuldade")
plt.ylabel("Frequência")
plt.title("Distribuição de Dificuldades")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Por meio do gráfico podemos perceber que a maioria das receitas é de fácil preparação

## 2 - tempo de preparo




```python
import json
import pathlib
import matplotlib.pyplot as plt

# Caminho da pasta com os arquivos JSON
data_dir = pathlib.Path("data")

# Contador
tempo_preparo_counter = {}

# Abrindo e processando os arquivos JSON
for json_file in data_dir.rglob("*.json"):
    with json_file.open("r", encoding="utf-8") as f:
        data = json.load(f)
        tempo_preparo = data.get("tempo_preparo")
        if tempo_preparo:
            tempo_preparo = tempo_preparo.lower()  # Converte para minúsculas
            if tempo_preparo in tempo_preparo_counter:
                tempo_preparo_counter[tempo_preparo] += 1
            else:
                tempo_preparo_counter[tempo_preparo] = 1



print("Contagem de tempos de preparo:")
for tempo, count in tempo_preparo_counter.items():
    print(f"{tempo}: {count}")

# Criando um gráfico de barras
plt.bar(tempo_preparo_counter.keys(), tempo_preparo_counter.values(), color="skyblue")
plt.xlabel("Tempo de Preparo")
plt.ylabel("Frequência")
plt.title("Distribuição de Tempos de Preparo")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Por meio dele podemos observar que grande parte das receitas levam até 1 hora para serem preparadas

## 3 - ingredientes

```python
from collections import Counter
import pandas as pd
import pathlib
import json
import re
import matplotlib.pyplot as plt

# Caminho da pasta com os arquivos JSON
data_dir = pathlib.Path("data")

# Lista de palavras a serem ignoradas (medidas, unidades, etc.)
stop_words = {
    'xicara', 'colher', 'copo', 'gramas', 'ml', 'g', 'kg', 'l', 'm', 'de', 'com', 'e', 'a', 'para', 'no', 'na',
   'sopa',  'temperos','1','2', '3', '4', 'chá', 'xícara', 'xícaras', 'em', 'lata', 'colheres', 'pó', 'ralado'
}

ingredientes_list = []

# Abrindo e processando os arquivos JSON
for json_file in data_dir.rglob("*.json"):
    with json_file.open("r", encoding="utf-8") as f:
        data = json.load(f)
        ingredientes = data.get("ingredientes", "")
        if ingredientes:
            # Separa os ingredientes por vírgulas, espaços e outros caracteres
            words = re.findall(r'\b\w+\b', ingredientes.lower())  # Considera apenas palavras
            # Filtra as palavras que não são ingredientes
            filtered_words = [word for word in words if word not in stop_words]
            ingredientes_list.extend(filtered_words)

# Contando as palavras mais comuns
word_counts = Counter(ingredientes_list)

# Obtendo as 10 palavras mais comuns
common_words = word_counts.most_common(10)

# Separando as palavras e suas contagens
words, counts = zip(*common_words)

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='skyblue')
plt.xlabel('Palavras')
plt.ylabel('Frequência')
plt.title('10 Palavras Mais Comuns em Ingredientes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("10 Palavras Mais Comuns em Ingredientes:")
for word, count in common_words:
    print(f"{word.capitalize()}: {count} vezes")
```

Como era de se esperar por se tratar de receitas de bolo, os 10 ingredientes mais utilizados são: leite, açúcar, fermento, ovos, farinha, trigo, óleo, coco, milho e margarina

# **Análise temporal**

---



A seguir foram verificados os tipos de bolo com mais receitas em cada mês, com o bolo de fubá se diferenciando dos demais e sendo o mais pesquisado em junho de 2021

```python
from collections import Counter
import pathlib
import json
import re
import matplotlib.pyplot as plt
from datetime import datetime

# Caminho da pasta com os arquivos JSON
data_dir = pathlib.Path("data")

# Lista de palavras a serem ignoradas
stop_words = {
    'bolo',  'de', 'com', 'e', 'a', 'para', 'no', 'na', 'receita', 'da','hora','fofinho','delicioso','simples','liquidificador','gostoso','muito','fácil'
}

monthly_words = {}

# Abrindo e processando os arquivos JSON
for json_file in data_dir.rglob("*.json"):
    with json_file.open("r", encoding="utf-8") as f:
        data = json.load(f)
        title = data.get("titulo", "")
        date_str = data.get("data", "")

        if title and date_str:
            # Extraindo o ano e mês da data
            date = datetime.fromisoformat(date_str)
            year_month = date.strftime("%Y-%m")  # Formato: "2024-12"

            # Separa as palavras do título
            words = re.findall(r'\b\w+\b', title.lower())  # Considera apenas palavras
            # Filtra as palavras que não são relevantes
            filtered_words = [word for word in words if word not in stop_words]

            # Adicionando as palavras ao dicionário por ano-mês
            if year_month not in monthly_words:
                monthly_words[year_month] = []
            monthly_words[year_month].extend(filtered_words)

# Contagem das palavras mais comuns por ano-mês
monthly_word_counts = {}
for year_month, words_list in monthly_words.items():
    word_counts = Counter(words_list)
    most_common_word = word_counts.most_common(1)
    if most_common_word:
        monthly_word_counts[year_month] = most_common_word[0]

# Ordenando os resultados por ano-mês
sorted_monthly_word_counts = sorted(monthly_word_counts.items())

# Exibindo os resultados
for year_month, (word, count) in sorted_monthly_word_counts:
    print(f"Em {year_month}, a palavra mais comum foi '{word}' com {count} ocorrências.")

months = [year_month for year_month, _ in sorted_monthly_word_counts]
words = [word for _, (word, _) in sorted_monthly_word_counts]
counts = [count for _, (_, count) in sorted_monthly_word_counts]

# Criando o gráfico
plt.figure(figsize=(12, 8))
plt.plot(months, counts, marker="o", linestyle="-", color="b")
for i, count in enumerate(counts):
    plt.text(months[i], count, f"{words[i]} ({count})", fontsize=8, ha="center", va="bottom")
plt.xlabel("Ano-Mês")
plt.ylabel("Frequência")
plt.title("Tipo de bolo mais comum em cada mês")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.grid(True)
plt.show()
```
